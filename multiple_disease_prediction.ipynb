```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
parkinsons_df = pd.read_csv(r"E:\User\Desktop\sham\DATA SCIENCE\PROJECTS\Project3 - Multiple Disease Prediction\files\parkinsons - parkinsons.csv")
feature = parkinsons_df.drop(columns=['name','status'], axis=1)
target = parkinsons_df['status']
# Set up the plot
plt.figure(figsize=(10, 6))

# Loop over all columns in the feature
for column in feature:
    sns.boxplot(y=parkinsons_df[column], label=column)
    plt.legend()
    plt.show()

# Show the plot
plt.title("Boxplot for all columns")
```

```python
#Find correlation between columns using heat map
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (20,20))
#Drop unwanted columns
feature = parkinsons_df.drop(columns=['name','status'])
sns.heatmap(feature.corr(), annot = True, fmt=".2f",linewidths=0.5)
```

```python
#Visualize Class distribution
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
parkinsons_df = pd.read_csv(r"E:\User\Desktop\sham\DATA SCIENCE\PROJECTS\Project3 - Multiple Disease Prediction\files\parkinsons - parkinsons.csv")
target = parkinsons_df['status']
# Plot a countplot to visualize class distribution
sns.countplot(x=target)
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
